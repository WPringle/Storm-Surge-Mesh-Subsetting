```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cmocean
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

## Reading from HDF5 file of input and output dataframes

```python
h5file = 'run_20210812_florence_multivariate_besttrack_250msubset_40members.h5'
```

```python
input_key = 'vortex_perturbation_parameters'
input_df = pd.read_hdf(h5file,input_key)
```

```python
display(input_df)
```

```python
output_key = 'zeta_max'
output_df = pd.read_hdf(h5file,output_key)
```

```python
display(output_df)
```

## Analysis 

### Get variablility of output 

```python
# just get standard deviation of the outputs
output_std = output_df.filter(regex='vortex*', axis=1).std(axis=1,skipna=True)
```

## Plotting

### Plot histograms of the input parameters (check look gaussian/uniform)

```python
# Plot variable distributions
# the histogram of the data
for variable in input_df:
    plt.hist(input_df[variable], bins='auto', density=True, alpha=0.75)
    plt.xlabel('Parameter')
    plt.ylabel('Probability')
    plt.title('Histogram of ' + variable)
    plt.grid(True)
    plt.show()
```

### Plot variability of maximum elevation

```python
#plot histogram
plt.hist(output_std, bins='auto', density=True, alpha=0.75)
plt.xlabel('Parameter')
plt.ylabel('Probability')
plt.title('Histogram of Maximum Elevation Variability')
plt.grid(True)
plt.show()
```

```python
# plot map
fig = plt.figure(figsize=(18,8))
ax = plt.axes(projection=ccrs.PlateCarree())
plt.scatter(output_df.x, output_df.y, 
            s=1,
            c=output_std,
            transform=ccrs.PlateCarree(),
            cmap=cmocean.cm.amp,
            vmin=0,vmax=1.6
)
plt.colorbar(ax=ax, shrink=.98,extend='max',label='STD [m]')
ax.coastlines()
# Add the gridlines
gl = ax.gridlines(color="black", linestyle="dotted", draw_labels=True, alpha=0.5)
gl.top_labels = None
gl.right_labels = None
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

plt.title("Maximum Elevation Variability from 40-member ensemble")
plt.show()
```
